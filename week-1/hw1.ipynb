# Exercise 2: Edge Extraction using Pytorch


To extract edges from a single gray-scale image, the below algorithm could be 
implemented. 

The Gaussian filter was introduced in the lecture
The Laplacian filter could be found in Gonzalez book, *Figure 3.39 d, page 129*

If you want to implement in on a 3-channel RGB input, split the input into 3 parts and apply **steps 3,4** separatedly on each channel, before concatenating the results.

**Edge Extraction Algorithm**
1. $x \gets read(path)$
2. $x_{preprocessed} \gets preprocess(x)$
3. $x_{smoothed} \gets x_{preprocessed} * filter_{gaussian}$
4. $egdes \gets x_{smoothed} * filter_{laplacian}$
4. $egdes \gets binarize(edges)$
